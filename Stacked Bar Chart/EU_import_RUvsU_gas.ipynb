```python
#This tells python to draw the graphs "inline" - in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
import seaborn as sns

import pylab
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("/content/Anson-Eva-Valerie/Stacked Bar Chart/data/quarterly_data 2024-10-03.xlsx")
df
```

```python
df['dates'] = pd.to_datetime(df['dates']).dt.strftime('%Y-%m-%d')

# set size,style
plt.figure(figsize=(10, 6))

# draw bar chart
plt.bar(df['dates'], df['Ukraine Gas Transit'], label='Ukraine Gas Transit')
plt.bar(df['dates'], df['Russia'],
        bottom=df['Ukraine Gas Transit'], label='Russia')

# add titles
plt.xlabel('Date')
plt.ylabel('Percentage (%)')
plt.title('Gas Import Dependency in Austria')
plt.legend()

# 45 degree rotate for nice to read
plt.xticks(rotation=45, ha='right')

# show the chart
plt.tight_layout()
plt.show()
```

1.整体依赖度显著下降
从左到右（即从 2021 年初到 2024 年底），柱子的总高度（蓝色+橙色的堆叠总量）呈逐步下降趋势。这说明从这两个渠道（乌克兰管道和俄罗斯）获得的天然气进口总量或所占份额在减少，也就是对这两个来源的依赖度持续降低

2.乌克兰管道气所占比例相对稳定
与橙色相比，蓝色部分的绝对值变化幅度相对较和俄罗斯战争，但整体看并没有像俄罗斯进口那样出现大幅波动。

3.对未来的启示
图中延伸到 2024 年后期，整体柱状高度仍然保持在相对较低水平，说明在可见的未来，对这两大来源的天然气依赖度保持下行或至少不再回到之前的高峰。需要在合同终止之前，找到新的能够代替的 天然气来源。
